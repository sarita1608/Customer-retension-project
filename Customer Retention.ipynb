```python
### Creatinf one Hot encodier for REcommend online retailers

data['recommend_amazon'] = data['Which of the Indian online retailer would you recommend to a friend?'].apply(lambda x :1 if (x.lower().find('amazon.in') >=0) else 0  )

data['recommend_flipkart'] = data['Which of the Indian online retailer would you recommend to a friend?'].apply(lambda x :1 if (x.lower().find('flipkart') >=0) else 0  )

data['recommend_myntra'] = data['Which of the Indian online retailer would you recommend to a friend?'].apply(lambda x :1 if (x.lower().find('myntra') >=0) else 0  )

data['recommend_paytm'] = data['Which of the Indian online retailer would you recommend to a friend?'].apply(lambda x :1 if (x.lower().find('paytm') >=0) else 0  )

data['recommend_snapdeal'] = data['Which of the Indian online retailer would you recommend to a friend?'].apply(lambda x :1 if (x.lower().find('snapdeal') >=0) else 0  )
```
